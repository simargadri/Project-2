```python
import pandas as pd
import numpy as np
from pathlib import Path
import yfinance as yf
```

```python
rbc = yf.Ticker("RY.TO")
rbc_df = rbc.history(period='max')
rbc_df = rbc_df.loc['2010':'2021']
rbc_df["Return"] = rbc_df['Close'].pct_change()
rbc_df["Return ewm"] = rbc_df["Return"].ewm(span=2, adjust=False).mean()
rbc_df["Vol change"] = rbc_df['Volume'].pct_change()
rbc_df["Volume Diff"] = rbc_df["Volume"].diff()
rbc_df["Chaikin"] = rbc_df["Volume Diff"].ewm(span=3, adjust=False).mean()
rbc_df["Chaikin Change"] = rbc_df["Chaikin"].pct_change()
rbc_df["High/Close"] = (rbc_df['High']/rbc_df['Close']) - 1
rbc_df["Low/Close"] = (rbc_df['Low']/rbc_df['Close']) - 1
rbc_df["High Change"] = rbc_df['High'].pct_change()
rbc_df["Low Change"] = rbc_df['Low'].pct_change()
rbc_df["Pos"] = np.where(rbc_df['Return'] > 0, '1', '0')
rbc_df["Spread"] = rbc_df['High'] - rbc_df['Low']
rbc_df["Spread Change"] = rbc_df['Spread'].pct_change()
rbc_df = rbc_df.dropna()
rbc_df.rename(columns = {"Close": "RBC Close"},  inplace = True)
rbc_df.head()
```

```python
vix = yf.Ticker("^VIX")
vix_df = vix.history(period='max')
vix_df = pd.DataFrame(vix_df["Close"])
vix_df["Vix return"] = vix_df['Close'].pct_change()
vix_df["Vix ewm"] = vix_df["Vix return"].ewm(span=2, adjust=False).mean()
vix_df = vix_df.loc['2010':'2021']
vix_df = vix_df.dropna()
vix_df.rename(columns = {"Close": "Vix Close"},  inplace = True)
vix_df.head()
```

```python
td = yf.Ticker("TD.TO")
td_df = td.history(period='max')
td_df = pd.DataFrame(td_df["Close"])
td_df["TD return"] = td_df['Close'].pct_change()
td_df = td_df.loc['2010':'2021']
td_df = td_df.dropna()
td_df.rename(columns = {"Close": "TD Close"},  inplace = True)
td_df.head()
```

```python
cad = yf.Ticker("CAD=X")
cad_df = cad.history(period='max')
cad_df = pd.DataFrame(cad_df["Close"])
cad_df["Cad return"] = cad_df['Close'].pct_change()
cad_df = cad_df.loc['2010':'2021']
cad_df = cad_df.dropna()
cad_df.rename(columns = {"Close": "CAD Close"},  inplace = True)
cad_df.head()
```

```python
tsx = yf.Ticker("^GSPTSE")
tsx_df = tsx.history(period='max')
tsx_df = pd.DataFrame(tsx_df["Close"])
tsx_df["Tsx return"] = tsx_df['Close'].pct_change()
tsx_df = tsx_df.loc['2010':'2021']
tsx_df = tsx_df.dropna()
tsx_df.rename(columns = {"Close": "TSX Close"},  inplace = True)
tsx_df.head()
```

```python
rbcmodel_df = pd.concat([rbc_df, vix_df, td_df, cad_df, tsx_df], axis=1, join="inner")
rbcmodel_df["Pos"] = rbcmodel_df["Pos"].shift(-1)
rbcmodel_df = rbcmodel_df.dropna()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
corrmat = rbcmodel_df.corr()
fig, heat = plt.subplots(figsize = (15, 10))
heat = sn.heatmap(corrmat, annot=True)
plt.show()
```

```python
rbc_df = rbcmodel_df[["Dividends", "High Change", "Low Change", "Vix ewm", "High/Close", "Low/Close", "Vix Close", "CAD Close", "Spread", "Spread Change", "Volume Diff", "Vol change", "Return ewm",  "Vix return", "TD return", "Tsx return", "Cad return", "Return", "Chaikin", "Chaikin Change", "Pos"]]
rbc_df
```
